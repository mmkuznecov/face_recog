```
import cv2
from imageai.Detection.Custom import CustomObjectDetection
from tqdm import tqdm_notebook as tqdm

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("/content/drive/My Drive/detection_model-ex-028--loss-0002.841.h5")
detector.setJsonPath("/content/drive/My Drive/detection_config (1).json")
detector.loadModel()
```

```
# fake_faces model on real faces
import os
no_faces = 0
ok = 0
too_much = 0
ln = len(os.listdir('/content/celebs_test'))
for image in tqdm(os.listdir('/content/celebs_test')):
  detections = detector.detectObjectsFromImage(input_image = '/content/celebs_test/' + image, output_image_path="holo2-detected.jpg")
  if len(detections) == 0:
    no_faces += 1
  elif len(detections) == 1:
    ok+=1
  else:
    too_much += 1
print('Ok at: ' + str((ok / ln) * 100) + ' %')
print('No at: ' + str((no_faces / ln) * 100) + ' %')
print('Too much at: ' + str((too_much / ln) * 100) + ' %')
```
