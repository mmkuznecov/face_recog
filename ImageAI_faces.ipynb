```
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
from imageai.Detection.Custom import CustomObjectDetection
import tqdm

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("/content/drive/My Drive/detection_model-ex-028--loss-0002.841.h5")
detector.setJsonPath("/content/drive/My Drive/detection_config (1).json")
detector.loadModel()

detections = detector.detectObjectsFromImage(input_image="/content/drive/My Drive/face_test.jpeg", output_image_path="holo2-detected.jpg")
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
a = cv2.imread('holo2-detected.jpg')
plt.imshow(a)
```
