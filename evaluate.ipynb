```python
from dlib import get_frontal_face_detector
import os
import cv2
```

```python
det = get_frontal_face_detector()

imgs = 0
faces = 0

dirs = os.listdir('celebs/')
dirs.remove('f_c.py')

for i in dirs:
    direct = 'celebs/' + i + '/'
    for j in os.listdir(direct):
        path = direct + j
        img = cv2.imread(path)
        d_faces = det(img, 1)
        imgs += 1
        if len(d_faces) == 1:
            faces += 1
            print('OK')
        elif len(d_faces) == 0:
            print(path)
            print('No face')
        else:
            print(path)
            print("too much faces")
            print(len(d_faces))
```

```python
cels = os.listdir('celebs/')
cels.remove('names')
cels.remove('f_c.py')

faces = 200
acc = 0


for i in cels:
    for j in os.listdir('celebs/' + i):
        image = cv2.imread('celebs/' + i + '/' + j)
        face = get_faces(image)
        encodings = get_embedding_list(face)
        names = compare_faces(encodings, known_encodings, known_names)
        name = names[0]
        if name == i:
            acc += 1
        else:
            print('celebs/' + i + '/' + j)
            print("Error")
print(acc / faces * 100.0)
```
