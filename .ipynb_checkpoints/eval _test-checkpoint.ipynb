```python
def get_name(image, th):
    boxes = get_boxes(image)
    #print(len(boxes))

    faces = get_faces(image)

    #print('Extracting faces')

    encodings = get_embedding_list(faces)

    #print('Extracting embeddings')

    names = compare_faces(encodings, known_encodings, known_names, th)
    name = names[0]
    return name
```

```python
import os
imgs = os.listdir('celebs/names')
for di in imgs:
    image = cv2.imread('celebs/names/' + di)
    name = get_name(image)
    if name == di.split('.')[0]:
        print('+')
    else:
        print(di)
```

```python
import os

ite = 0
plus = 0

dirs = os.listdir('celebs')
dirs.remove('names')
for i in dirs:
    for im in os.listdir('celebs/' + i):
        ite += 1
        image = cv2.imread('celebs/' + i + '/' + im)
        name = get_name(image)
        if name == im.split('.')[0][:-1]:
            plus += 1
        else:
            print('ERROR: PRINTED ' + name +' EXPECTED ' + im.split('.')[0][:-1])
            print(im)
print("Accuracy: " + str((plus / ite) * 100))
```

```python
import os

ma1 = 0
ma2 = 0
ite = 0
plus = 0

dirs = os.listdir('celebs')
dirs.remove('names')

dist = 0
for k in range(100):
    ma2 = ma1
    dist += 1
    for i in dirs:
        for im in os.listdir('celebs/' + i):
            ite += 1
            image = cv2.imread('celebs/' + i + '/' + im)
            name = get_name(image, dist)
            if name == im.split('.')[0][:-1]:
                plus += 1
    ma1 = plus / ite
    print("MA1: " + str(ma1))
    if ma2 < ma1:
        print('MAX at: ' + str(dist))
```
